```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
daily_arrivals = np.array([8, 10, 15, 9, 13, 11, 14, 7, 16, 10, 12, 9, 15, 11, 13, 14, 8, 12, 10])

# Mean
mean_arrivals = np.mean(daily_arrivals)

# Population variance (ddof=0)
pop_variance = np.var(daily_arrivals, ddof=0)

# Sample variance (ddof=1)
sample_variance = np.var(daily_arrivals, ddof=1)

# Sample standard deviation (ddof=1)
std_dev = np.std(daily_arrivals, ddof=1)

print(f"Mean: {mean_arrivals:.2f}")
print(f"Population Variance: {pop_variance:.2f}")
print(f"Sample Variance: {sample_variance:.2f}")
print(f"Sample Standard Deviation: {std_dev:.2f}")
print("\nComparison:")
print(f"Sample variance is slightly larger than population variance because sample variance divides by (n-1) instead of n, providing an unbiased estimator.")
```

```python
# 2. Random Number Generation
np.random.seed(42)
generated_data = np.random.normal(mean_arrivals, std_dev, 20)

# Plotting histograms side by side
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.hist(daily_arrivals, bins=7, alpha=0.7, color='blue', edgecolor='black')
plt.title('Original Exam Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')

plt.subplot(1,2,2)
plt.hist(generated_data, bins=7, alpha=0.7, color='orange', edgecolor='black')
plt.title('Generated Exam Scores (Normal Distribution)')
plt.xlabel('Score')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Set random seed for reproducibility (optional)
np.random.seed(42)

# Number of simulations
num_simulations = 100000

# Simulate arrivals using uniform distribution from 7 to 16
simulated_arrivals = np.random.uniform(low=7, high=16, size=num_simulations)

# Estimate the probability that arrivals > 15
prob_more_than_15 = np.mean(simulated_arrivals > 15)

# Print the result
print(f"Estimated Probability (Arrivals > 15): {prob_more_than_15:.4f}")
```
