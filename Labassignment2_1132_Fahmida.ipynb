```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
observed = np.array([45, 35, 25])
```

```python
total = np.sum(observed)
expected = np.full_like(observed, total / len(observed))  # Uniform distribution
```

```python
chi_square_stat = np.sum(((observed - expected) ** 2) / expected)
```

```python
df = len(observed) - 1
```

```python
p_value = 1 - stats.chi2.cdf(chi_square_stat, df)  # p-value
alpha = 0.05  # Significance level
critical_value = stats.chi2.ppf(1 - alpha, df)  # Critical value
```

```python
print("Assignment 1: Chi-Square Goodness of Fit (Uniform Distribution)")
print(f"Observed: {observed}")
print(f"Expected: {expected}")
print(f"Chi-square statistic = {chi_square_stat:.3f}")
print(f"Degrees of freedom = {df}")
print(f"Critical value (0.05) = {critical_value:.3f}")
print(f"P-value = {p_value:.4f}")
```

```python
if chi_square_stat > critical_value:
    print("Reject H0 → Observed data does NOT fit a uniform distribution.")
else:
    print("Fail to reject H0 → Observed data fits a uniform distribution.")
```

```python
x = np.arange(1, len(observed) + 1)
plt.bar(x, observed, width=0.4, label='Observed', alpha=0.7)
plt.bar(x + 0.4, expected, width=0.4, label='Expected', alpha=0.7)
plt.xticks(x + 0.2, ['Cat1', 'Cat2', 'Cat3'])
plt.xlabel('Categories')
plt.ylabel('Frequencies')
plt.title('Assignment 1: Observed vs Expected Frequencies')
plt.legend()
plt.show()
```

```python
# Assignment 2
observed = np.array([50, 30, 15, 5])
expected_proportion = np.array([0.6, 0.25, 0.10, 0.05])
total = np.sum(observed)
expected = expected_proportion * total

chi_square_stat = np.sum(((observed - expected) ** 2) / expected)
df = len(observed) - 1
alpha = 0.05
critical_value = stats.chi2.ppf(1 - alpha, df)
p_value = 1 - stats.chi2.cdf(chi_square_stat, df)

print("\nAssignment 2: Chi-Square Goodness of Fit (Custom Expected Distribution)")
print(f"Observed: {observed}")
print(f"Expected: {expected}")
print(f"Chi-square statistic = {chi_square_stat:.3f}")
print(f"Degrees of freedom = {df}")
print(f"Critical value (0.05) = {critical_value:.3f}")
print(f"P-value = {p_value:.4f}")

if chi_square_stat > critical_value:
    print("Reject H0 → Observed data does NOT fit expected distribution.")
else:
    print("Fail to reject H0 → Observed data fits expected distribution.")

# Graph
x = np.arange(1, len(observed) + 1)
plt.bar(x, observed, width=0.4, label='Observed', alpha=0.7)
plt.bar(x + 0.4, expected, width=0.4, label='Expected', alpha=0.7)
plt.xticks(x + 0.2, ['A', 'B', 'C', 'D'])
plt.xlabel('Categories')
plt.ylabel('Frequencies')
plt.title('Assignment 2: Observed vs Expected Frequencies')
plt.legend()
plt.show()
```
